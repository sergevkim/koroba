```python
import numpy as np
import open3d as o3d
import plotly.graph_objects as go

from koroba.utils import Box, Visualizer


cloud = o3d.io.read_point_cloud("../data/bunny.pcd")
cloud.estimate_normals(search_param=o3d.geometry.KDTreeSearchParamHybrid(radius=0.1, max_nn=30))

points = np.asarray(cloud.points)
horizontal_bbox = Box.estimate_horizontal_bounding_box(points)
g2 = Visualizer.get_geometry(horizontal_bbox)

Visualizer.draw_geometries([cloud] + g2)

```
